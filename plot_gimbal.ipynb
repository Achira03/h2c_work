```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV
df = pd.read_csv("gimbal_data_20250715_113302.csv")

# Plot  pitch_angle  yaw_angle 
plt.figure(figsize=(15, 6))
plt.plot(df['time'], df['pitch_angle'], label='Pitch Angle')
plt.plot(df['time'], df['yaw_angle'], label='Yaw Angle')

plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.title('Gimbal Pitch & Yaw Angle vs Time')
plt.legend()
plt.grid(True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  CSV
file_path = "gimbal_tof_data_2.csv"  #  path 
df = pd.read_csv(file_path)

#  tof1  0 ( plot )
valid_data = df[df['tof1'] > 0]

#  yaw_ground_angle 
angles_rad = np.deg2rad(valid_data['yaw_ground_angle'].values)
distances = valid_data['tof1'].values

#  7 cm (70 mm)
distances = distances + 70

#  x, y ()
x = distances * np.cos(angles_rad)
y = distances * np.sin(angles_rad)

#  XY  ( y,  x)
plt.figure(figsize=(8, 8))
plt.scatter(y, x, c='blue', label='TOF1 (+70mm)', marker='x')
plt.xlabel("Y (mm)")
plt.ylabel("X (mm)")
plt.title("gimbal TOF1 Distance (Y Horizontal / X Vertical)")
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid(True)
plt.show()
```
