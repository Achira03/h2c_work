```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV
# 
df = pd.read_csv("tof_data_raw.csv") # <--  CSV 
df['timestamp'] = pd.to_datetime(df['timestamp'])

#  ToF  ( tof1)
tof_column = 'tof1'

#  DataFrame 
df_filtered = df.copy()

# --- Moving Average Filter ---
window_size_ma = 5
df_filtered[f'{tof_column}_MA_w{window_size_ma}'] = df[tof_column].rolling(window=window_size_ma, center=True).mean()

# ---  ---

#  1:  (Original ToF1)
plt.figure(figsize=(15, 7)) # 
plt.plot(df['timestamp'], df[tof_column], label='Original ToF1', color='gray', alpha=0.9)
plt.axhline(y=400, color='red', linestyle='--', label='Stop Threshold (400 mm)')
plt.title(f'Original ToF1 Distance', fontsize=16)
plt.xlabel('Timestamp', fontsize=12)
plt.ylabel('Distance (mm)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tick_params(axis='x', rotation=45) #  timestamp 
plt.tight_layout() # 
plt.show()

#  2:  Moving Average Filter
plt.figure(figsize=(15, 7)) # 
plt.plot(df_filtered['timestamp'], df_filtered[f'{tof_column}_MA_w{window_size_ma}'], label=f'Moving Average (Window={window_size_ma})', linestyle='-', color='blue')
plt.axhline(y=400, color='red', linestyle='--', label='Stop Threshold (400 mm)')
plt.title(f'ToF Distance after Moving Average Filter (Window {window_size_ma})', fontsize=16)
plt.xlabel('Timestamp', fontsize=12)
plt.ylabel('Distance (mm)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tick_params(axis='x', rotation=45) #  timestamp 
plt.tight_layout() # 
plt.show()

print(f" ToF1 (Original)  ToF1 (Moving Average Filter) .")
```
