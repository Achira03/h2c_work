```python
import pandas as pd
import matplotlib.pyplot as plt

# อ่านไฟล์ CSV
# ตรวจสอบให้แน่ใจว่าชื่อไฟล์ถูกต้องตามไฟล์ที่คุณสร้าง
df = pd.read_csv("tof_data_raw.csv") # <-- แก้ไขชื่อไฟล์ CSV ให้ถูกต้อง
df['timestamp'] = pd.to_datetime(df['timestamp'])

# กำหนดคอลัมน์ ToF ที่ต้องการวิเคราะห์ (เฉพาะ tof1)
tof_column = 'tof1'

# สร้าง DataFrame สำหรับเก็บข้อมูลที่กรองแล้ว
df_filtered = df.copy()

# --- Moving Average Filter ---
window_size_ma = 5
df_filtered[f'{tof_column}_MA_w{window_size_ma}'] = df[tof_column].rolling(window=window_size_ma, center=True).mean()

# --- เริ่มสร้างกราฟแยกกัน ---

# กราฟที่ 1: ข้อมูลดิบ (Original ToF1)
plt.figure(figsize=(15, 7)) # ปรับขนาดกราฟให้ใหญ่ขึ้น
plt.plot(df['timestamp'], df[tof_column], label='Original ToF1', color='gray', alpha=0.9)
plt.axhline(y=400, color='red', linestyle='--', label='Stop Threshold (400 mm)')
plt.title(f'Original ToF1 Distance', fontsize=16)
plt.xlabel('Timestamp', fontsize=12)
plt.ylabel('Distance (mm)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tick_params(axis='x', rotation=45) # หมุนป้าย timestamp เพื่อให้อ่านง่ายขึ้น
plt.tight_layout() # ปรับเลย์เอาต์ให้อัตโนมัติ
plt.show()

# กราฟที่ 2: ข้อมูลที่ผ่าน Moving Average Filter
plt.figure(figsize=(15, 7)) # ปรับขนาดกราฟให้ใหญ่ขึ้น
plt.plot(df_filtered['timestamp'], df_filtered[f'{tof_column}_MA_w{window_size_ma}'], label=f'Moving Average (Window={window_size_ma})', linestyle='-', color='blue')
plt.axhline(y=400, color='red', linestyle='--', label='Stop Threshold (400 mm)')
plt.title(f'ToF Distance after Moving Average Filter (Window {window_size_ma})', fontsize=16)
plt.xlabel('Timestamp', fontsize=12)
plt.ylabel('Distance (mm)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tick_params(axis='x', rotation=45) # หมุนป้าย timestamp เพื่อให้อ่านง่ายขึ้น
plt.tight_layout() # ปรับเลย์เอาต์ให้อัตโนมัติ
plt.show()

print(f"กราฟ ToF1 (Original) และ ToF1 (Moving Average Filter) แสดงแยกกันแล้ว.")
```
