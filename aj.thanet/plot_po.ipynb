```python
# kp_0.5

import pandas as pd
import matplotlib.pyplot as plt

# อ่านข้อมูลจากไฟล์ CSV
df = pd.read_csv("position_log.csv")

# สร้างกราฟ
plt.figure(figsize=(10, 5))
plt.plot(df["timestamp"], df["x"], marker='o', color='red', label='Position X (m)')
plt.title("Position X for RoboMaster S1")
plt.xlabel("Timestamp")
plt.ylabel("Pos X (M)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
